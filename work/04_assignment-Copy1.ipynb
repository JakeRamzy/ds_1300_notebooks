```python
Finaloil['Date_dt'] = pd.to_datetime(Finaloil['Record Date'],errors='raise')
Finaloil['Date_dt']
Finaloil['DateInt']=Finaloil['Date_dt'].astype(int)/10e10
```

```python
Finaloil=Finaloil.sort_values(by='Date_dt')
```
